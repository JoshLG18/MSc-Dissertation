```python
# import necessary modules
from data_processing import load_data, clean_news_data

# run the news data processing
news_data = load_data('../data/raw/NYT_Top_Daily_Articles.csv')
cleaned_news_data = clean_news_data(news_data)
print(cleaned_news_data.head())
```

```python
# Plot sentiment over time (e.g., daily or monthly average sentiment)
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 6))
plt.plot(cleaned_news_data['Date'], cleaned_news_data['sentiment'], marker='o', linestyle='-', alpha=0.7)
plt.title('Sentiment Score Over Time')
plt.xlabel('Date')
plt.ylabel('Average Daily Sentiment Score')
plt.grid(True)
plt.tight_layout()
plt.show()

# Optionally, print summary statistics
print(cleaned_news_data['sentiment'].describe())
```
